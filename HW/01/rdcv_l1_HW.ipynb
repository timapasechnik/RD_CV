```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('whale.jpg')
img = cv2.cvtColor( img,  cv2.COLOR_BGR2RGB)
rows,cols, chans = img.shape

plt.rcParams['figure.figsize'] = [ rows / 90 , cols / 90 ]
plt.imshow( img )
```

```python
red, green, blue = cv2.split(img)

plt.subplot(131), plt.imshow( red, cmap = "gray" )
plt.subplot(132), plt.imshow( green, cmap = "gray" )
plt.subplot(133), plt.imshow( blue, cmap = "gray" )
```

```python
img1 = cv2.merge ([red, green, blue])
img2 = cv2.merge ([red, blue, green])
img3 = cv2.merge ([green, red, blue])
img4 = cv2.merge ([blue, green, red])

out1 = np.hstack( [img1, img2])
out2 = np.hstack( [img3, img4])
final_out = np.vstack([out1,out2])

plt.imshow(final_out)
plt.axis(False)
```
