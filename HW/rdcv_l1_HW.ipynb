```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('img_002.png')
img = cv2.cvtColor( img,  cv2.COLOR_BGR2RGB)
rows,cols, chans = img.shape
```

```python
plt.rcParams['figure.figsize'] = [ rows / 90 , cols / 90 ]
plt.imshow( img )
```

```python
red, green, blue = cv2.split(img)

#quiet interesting result if we show one chanel of img
#can't understand  why we have such result - expected result was a grayscale picture like red is a 2d array.
#after reads of references found an answer)) 
plt.subplot(131), plt.imshow( red, cmap = "gray" )
plt.subplot(132), plt.imshow( green, cmap = "gray" )
plt.subplot(133), plt.imshow( blue, cmap = "gray" )
```

```python
img1 = cv2.merge ([red, green, blue])
img2 = cv2.merge ([red, blue, green])
img3 = cv2.merge ([green, red, blue])
img4 = cv2.merge ([blue, green, red])

out1 = np.hstack( [img1, img2])
out2 = np.hstack( [img3, img4])
final_out = np.vstack([out1,out2])

plt.imshow(final_out)
plt.axis(False)  
```

```python
#some fun with arrays to achive abaration effect look like )) 
nblue = blue
nblue [ :, 5: ] = blue[:, :-5 ]
ngreen = green
ngreen [ 5:, : ] = green [:-5, : ]

new_img = cv2.merge ( [red, green, nblue ] )
plt.imshow(new_img)
```
